# Ultimate Pig - An Analysis of Play Styles

#### The purpose of this exercise is to analyze three different styles of play for Pig Solitaire. This is accomplished through use of three Python Objects representing the play styles. The resulting data was culled from a number of simulations for each of the three playstyles.

The three play styles are:
1. 'Basic' - This player type will roll the die one time and one time only. This player quits ("holds") after the first roll.
2. 'Random' - This player type will choose randomly between rolling the die again after the first roll and for each subsequent turn.
3. 'Gambler' - This player type is unlikely to hold and will always roll until their score for the game is at 70. From that point, the Gambler has a 50/50 shot of pressing his luck or holding.

The rules of Pig Solitaire are:
>In 7 turns, you are attempting to get the highest score possible. 
>Each turn, you have two choices:

>Roll. Roll a six-sided die. If it comes up one, your turn is over and you add nothing to your score. If it comes up two through six, you add that number to your turn total and choose again.
>Hold. If you hold, you add the turn total to your score.

## Basic Code:

```python
%matplotlib inline
import random
import math
import statistics as st
import numpy as np
import matplotlib.pyplot as plt
import seaborn
```

```python
class Player:

    def __init__(self):
        self.game_score = 0
        self.turns = 0

    def roll_a_die(self):
        return random.randint(1, 6)

    def roll_again(self):
        return False

    def turn(self):
        roll_results_list = []
        while True:
            roll = self.roll_a_die()
            if roll == 1:
                roll_results_list = [0]
                return sum(roll_results_list)
                break
            else:
                roll_results_list.append(roll)
                roll_or_hold = self.roll_again()
                if roll_or_hold == False:
                    return sum(roll_results_list)
                    break
                else:
                    roll_results_list.append(roll)
                    continue
        self.game_score.append(roll_results_list)
        return roll_results_list

    def game(self, roll_results_list):
        while self.turns <= 6:
            roll_results_list = self.turn()
            self.turns += 1
            self.game_score += roll_results_list
        return self.game_score

class RandomPlayer(Player):

    def roll_again(self):
        return random.choice([True, False])

class GamblerPlayer(Player):

    def roll_again(self):
        if self.game_score < 70:
            return random.choice([True, False])
```

### Now, the actual trial code:

```python
trial_results = []
for _ in range(50):
    roll_results_list = []
    player = Player()
    player.game(roll_results_list)
    trial_results.append(player.game(roll_results_list))
```

And, the results:

```python
print(trial_results)
```

The above data is for the base class. Remember, this player plays extremely cautiously and will always stop after one roll, meaning that their ceiling for any given round is 6 (one roll of a six-sided die) and thus their ceiling for a full seven round (standard) game should be 42.

```python
x1 = range(len(trial_results))
y1 = trial_results
plt.plot(y1)
plt.scatter(range(len(trial_results)), y1)
plt.xticks([10, 20, 30, 40, 50], x1)
plt.xlabel("Trial Number")
plt.ylabel("Resulting Score")
plt.title("Base Player Scores over Trials")
plt.show()
```

The resulting graph suggests a larger variation in scores than actually exists. Please note the scale of the Y-Axis.

Now, we run the simulation the same number of times for the random playter type. This player will decide to continue on after the first roll and each subsequent roll by flip of a coin, basically.

```python
trial_results = []
for _ in range(50):
    roll_results_list = []
    player = RandomPlayer()
    player.game(roll_results_list)
    trial_results.append(player.game(roll_results_list))
```

```python
print(trial_results)
```

From the results, an eyeball test suggests he routinely wins with a higher score than the base player who only rolls once.

```python
x1 = range(len(trial_results))
y1 = trial_results
plt.plot(y1)
plt.scatter(range(len(trial_results)), y1)
plt.xticks([10, 20, 30, 40, 50], x1)
plt.xlabel("Trial Number")
plt.ylabel("Resulting Score")
plt.title("Random Player Scores over Trials")
plt.show()
```

The visual representation of the data supports this assertion; note that the Y-Axis is on a larger scale and thus the magnitude is greater.

We now turn to the Gambler player type simulations.

```python
trial_results = []
for _ in range(50):
    roll_results_list = []
    player = GamblerPlayer()
    player.game(roll_results_list)
    trial_results.append(player.game(roll_results_list))
```

```python
print(trial_results)
```

```python
x1 = range(len(trial_results))
y1 = trial_results
plt.plot(y1)
plt.scatter(range(len(trial_results)), y1)
plt.xticks([10, 20, 30, 40, 50], x1)
plt.xlabel("Trial Number")
plt.ylabel("Resulting Score")
plt.title("Gambler Player Scores over Trials")
plt.show()
```

Based on these outputs, it does not appear that the Gambler's winning strategy is any more valid than the Random player's strategy. Let's look at some more statistics.

```python
class_results = []
for _ in range(100):
    trial_results = []
    for _ in range(50):
        roll_results_list = []
        player = Player()
        player.game(roll_results_list)
        trial_results.append(player.game(roll_results_list))
    class_results.append(trial_results)
```

Here we are setting up a larger data set; 100 trials of 50 simulations.

```python
list_of_averages = []
for collection in class_results:
    list_of_averages.append((sum(collection) / len(collection)))
```

And above, we compile a list of averages for each set of 50 simulations in the 100 trials. The list of averages:

```python
print(list_of_averages)
```

```python
plt.hist(list_of_averages)
mean = st.mean(list_of_averages)
stdev = st.stdev(list_of_averages)
ymin, ymax = plt.ylim()

plt.vlines(mean, ymin, ymax, color='yellow')
plt.vlines([mean - 2 * stdev, mean - stdev, mean + stdev, mean + 2 * stdev], ymin, ymax, color='blue', linestyles='dashed')
plt.show()
```

The above graph displays the distributions of outcomes over the trials. The yellow line represents the mean of the data set, or the average score returned over all of the trials by the base player type. The dotted lines represent standard deviations from that mean.

Now, let's repeat this process with the Random player type:

```python
class_results = []
for _ in range(100):
    trial_results = []
    for _ in range(50):
        roll_results_list = []
        player = RandomPlayer()
        player.game(roll_results_list)
        trial_results.append(player.game(roll_results_list))
    class_results.append(trial_results)
```

```python
list_of_averages = []
for collection in class_results:
    list_of_averages.append((sum(collection) / len(collection)))
```

```python
print(list_of_averages)
```

These results indicate a much higher average score, although the distribution is more uneven. By picking purely randomly whether to roll the die a second or more times or not, the player has a higher chance of obtaining a highe score.

Finally, let's look at the Gambler and see how he fares:

```python
class_results = []
for _ in range(100):
    trial_results = []
    for _ in range(50):
        roll_results_list = []
        player = GamblerPlayer()
        player.game(roll_results_list)
        trial_results.append(player.game(roll_results_list))
    class_results.append(trial_results)
```

```python
list_of_averages = []
for collection in class_results:
    list_of_averages.append((sum(collection) / len(collection)))
```

```python
print(list_of_averages)
```

```python
plt.hist(list_of_averages)
mean = st.mean(list_of_averages)
stdev = st.stdev(list_of_averages)
ymin, ymax = plt.ylim()

plt.vlines(mean, ymin, ymax, color='yellow')
plt.vlines([mean - 2 * stdev, mean - stdev, mean + stdev, mean + 2 * stdev], ymin, ymax, color='blue', linestyles='dashed')
plt.show()
```

The mean for the Gambler is actually lower than for the Random player type. This means it is actually a better strategy to flip a coin on any given turn than to play aggressively and roll with a "hot hand" up until a threshold of 70 points. The lower occurrence of high averages on the right side of the graph suggest that this play style results in many more "busts", by which I mean rolling a one after having achieved some success on previous rolls.

In conclusion, amongst these three player types (extremely cautious Base player, completely random and arbitrary Random player, and aggressive thrill-seeking Gambler type), it is better to play a random playstyle characterized by a flip of the coin to decide whether to roll again or not on any given turn.
